```python
import icu

def get_script_of_char(char):
    sr = icu.ScriptRun(char)
    if sr.next():  # Avanza al primer "run"
        code = sr.getScriptCode()
        # Tabla reducida de los scripts más comunes (puedo darte la completa si quieres)
        script_names = {
            0: "Common",
            1: "Inherited",
            2: "Arabic",
            3: "Armenian",
            4: "Bengali",
            5: "Bopomofo",
            6: "Cherokee",
            8: "Cyrillic",
            9: "Devanagari",
            10: "Ethiopic",
            11: "Georgian",
            12: "Greek",
            13: "Gujarati",
            14: "Gurmukhi",
            15: "Han",
            16: "Hangul",
            17: "Hebrew",
            18: "Hiragana",
            20: "Katakana",
            25: "Latin",
        }
        return script_names.get(code, f"Unknown({code})")
    return None


# Ejemplos
for c in ['あ', 'ア', 'A', 'Ж', 'م', 'क']:
    print(f"{c}: {get_script_of_char(c)}")
```
